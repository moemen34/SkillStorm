# Highest grossing movie for each genre

```python
# using exploded from b4 (highest rating for genre)
df4 = pd.read_csv("c:/Users/moeme/OneDrive/Desktop/Training/Python/cloned/20240617-DE-TS-LectureMaterials/Python/Data/imdb_top_1000.csv")
df4['Genre'] = df4['Genre'].str.split(', ')
df4_exploded = df4.explode('Genre')

df4_exploded.drop(columns = ['Poster_Link', 'Released_Year', 'Certificate', 'Runtime', 'IMDB_Rating', 'Overview', 'Meta_score', 'Director', 'Star1', 'Star2', 'Star3', 'Star4', 'No_of_Votes'], inplace=True)

df4_exploded.info()

#df4.info()
```

# Aggregate gross rev for all 1000 movies

```python
df7 = pd.read_csv("c:/Users/moeme/OneDrive/Desktop/Training/Python/cloned/20240617-DE-TS-LectureMaterials/Python/Data/imdb_top_1000.csv")

df7.drop(columns = ['Poster_Link', 'Director', 'Released_Year', 'Certificate', 'Genre', 'Runtime', 'IMDB_Rating', 'Overview', 'Meta_score', 'Star1', 'Star2', 'Star3', 'Star4', 'No_of_Votes'], inplace=True)


# Gross is an object, we need to convert it to aa numeric type

# remove the ',' because converting throws error
df7['Gross'] = df7['Gross'].str.replace(',', '').astype(float)

#convert
df7['Gross'] = pd.to_numeric(df7['Gross'], errors='coerce')

df7.head()
```

```python
non_numeric_rows = df7[df7['Gross'].isnull()]

# Display rows with non-numeric values
print("Rows =that are not numeric:")
print(non_numeric_rows.shape[0])
non_numeric_rows.head()
```

```python
# find the average gross and place it for rows that don't have an average

# average/mean
average_gross = df7['Gross'].mean()

average_gross

# replace NaN with average

#df7[df7['Gross'].isnull()]
df7['Gross'].fillna(average_gross, inplace=True)

# Chack if there's more nan
df7[df7['Gross'].isnull()]
```
